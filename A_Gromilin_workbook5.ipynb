# Python

1. –î–∞–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –¢—Ä–µ–±—É–µ—Ç—Å—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ
—Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω—ã–º –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º, –∫—Ä–∞—Ç–Ω—ã–º –ø—è—Ç–∏.

```python
a = int(input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à–µ —á–∏—Å–ª–æ: "))
if (a%5==0) and (100<=a<=999):
  print(f"–ß–∏—Å–ª–æ {a} —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –≤—Å–µ–º —É—Å–ª–æ–≤–∏—è–º")
else:
  print(f"–ß–∏—Å–ª–æ {a} –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —É—Å–ª–æ–≤–∏—è–º")
```

2. –î–∞–Ω—ã 3 —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞. –¢—Ä–µ–±—É–µ—Ç—Å—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –µ—Å—Ç—å –ª–∏ —Å—Ä–µ–¥–∏ —ç—Ç–∏—Ö —á–∏—Å–µ–ª
—Ö–æ—Ç—è –±—ã –¥–≤–∞ —á–µ—Ç–Ω—ã—Ö.

```python
def main(a, b, c):
    d = sum(1 for e in (a, b, c) if e % 2 == 0)
    return d >= 2

a = int(input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ: "))
b = int(input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ: "))
c = int(input("–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–µ—Ç—å–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ: "))

if main(a, b, c):
    print("–°—Ä–µ–¥–∏ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –µ—Å—Ç—å —Ö–æ—Ç—è –±—ã –¥–≤–∞ —á–µ—Ç–Ω—ã—Ö.")
else:
    print("–°—Ä–µ–¥–∏ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ—Ç —Ö–æ—Ç—è –±—ã –¥–≤—É—Ö —á–µ—Ç–Ω—ã—Ö.")
```
