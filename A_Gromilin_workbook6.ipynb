1. –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ —Å–ª–æ–≤–æ. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ
–æ–Ω–æ —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º. –ü–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å–ª–æ–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ
–æ–¥–∏–Ω–∞–∫–æ–≤–æ —á–∏—Ç–∞–µ—Ç—Å—è –∫–∞–∫ —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ, —Ç–∞–∫ –∏ —Å–ø—Ä–∞–≤–∞ –Ω–∞–ª–µ–≤–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –≤
–∞–Ω–≥–ª–∏–π—Å–∫–æ–º —è–∑—ã–∫–µ —Ç–∞–∫–∏–º–∏ —Å–ª–æ–≤–∞–º–∏ —è–≤–ª—è—é—Ç—Å—è "radar" –∏ "racecar".

```python
a = str(input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à–µ —Å–ª–æ–≤–æ: "))
flag = False
for i in range(len(a)):
  if a[:i] + a[i+1:] == (a[:i] + a[i+1:])[::-1]:
      print(f"–°–ª–æ–≤–æ '{a}' —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º")
      flag = True
      break
if flag == False:
    print(f"–°–ª–æ–≤–æ '{a}' –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º")
```
