1. Пользователь вводит произвольное слово. Необходимо проверить, что
оно является палиндромом. Палиндромом называется слово, которое
одинаково читается как слева направо, так и справа налево, например, в
английском языке такими словами являются "radar" и "racecar".

```python
a = str(input("Введите ваше слово: "))
flag = False
for i in range(len(a)):
  if a[:i] + a[i+1:] == (a[:i] + a[i+1:])[::-1]:
      print(f"Слово '{a}' является палиндромом")
      flag = True
      break
if flag == False:
    print(f"Слово '{a}' не является палиндромом")
```

7.  Если объект range (диапазон) передать встроенной в Python функции list(), то она преобразует его к списку. Создайте, таким образом, список с элементами от 0 до 100 и шагом 17.

```python
a = list(range(0,100))
b = []
i=0
while i < 100:
  b.insert(i, a[i])
  i+=17
  print(b)
```

8. Напишите программу, которая заполняет список пятью словами, введенными с клавиатуры, измеряет длину каждого слова и добавляет полученное значение в другой список. Например, список слов – ['yes', 'no', 'maybe', 'ok', 'what'], список длин – [3, 2, 5, 2, 4]. Оба списка должны выводиться на экран.

```python
words = []
lengths = []
for i in range(5):
    word = input("Введите ваше слово: ")
    words.append(word)
    lengths.append(len(word))
print("Список слов:", words)
print("Список длин:", lengths)
```

9. Даны два множества X и Y. Найти:

1) их объединение

2) их пересечение

3) элементы множества A, которых нет во множестве B

```python
X = {1, 2, 3, 4, 5}
Y = {3, 4, 5, 6, 7}

print("Объединение множеств X и Y:", X | Y)
print("Пересечение множеств X и Y:", X & Y)
print("Элементы множества A, которых нет во множестве B:", X ^ Y)
```
