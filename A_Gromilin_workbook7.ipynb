1. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞
(–¥–µ–ª–∏–º–æ–µ –∏ –¥–µ–ª–∏—Ç–µ–ª—å) –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ü–µ–ª—É—é —á–∞—Å—Ç—å –æ—Ç –¥–µ–ª–µ–Ω–∏—è –∏ –æ—Å—Ç–∞—Ç–æ–∫.

```python
x=int(input("–í–≤–µ–¥–∏—Ç–µ –¥–µ–ª–∏–º–æ–µ"))
y=int(input("–í–≤–µ–¥–∏—Ç–µ –¥–µ–ª–∏—Ç–µ–ª—å"))
def f(x,y):
    return(x // y,x % y)
print(f"–¶–µ–ª–∞—è —á–∞—Å—Ç—å –∏ –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è: {f(x,y)}")
```

2.–û–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é Sign(X) —Ü–µ–ª–æ–≥–æ —Ç–∏–ø–∞, –≤–æ–∑–≤—Ä–∞—â–∞—é—â—É—é –¥–ª—è
–≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ X —Å–ª–µ–¥—É—é—â–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è: ‚àí1, –µ—Å–ª–∏ X < 0; 0, –µ—Å–ª–∏ X =
0; 1, –µ—Å–ª–∏ X > 0.

```python
x=int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ"))
def Sign(x):
    if x < 0:
        return -1
    if x == 0:
        return 0
    if x >0:
        return 1
print(f"–ß–∏—Å–ª–æ {x} –ø–æ—Å–ª–µ —Ñ—É–Ω–∫—Ü–∏–∏ Sign(x) —Ä–∞–≤–Ω–æ: {Sign(x)}")
```

3. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≤–µ—Ä–∫—É –Ω–∞ —Ç–æ, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º. –ü–∞–ª–∏–Ω–¥—Ä–æ–º
‚Äî —ç—Ç–æ —Å–ª–æ–≤–æ –∏–ª–∏ —Ñ—Ä–∞–∑–∞, –∫–æ—Ç–æ—Ä—ã–µ –æ–¥–∏–Ω–∞–∫–æ–≤–æ —á–∏—Ç–∞—é—Ç—Å—è —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ –∏
—Å–ø—Ä–∞–≤–∞ –Ω–∞–ª–µ–≤–æ.

```python
a = str(input("–í–≤–µ–¥–∏—Ç–µ –≤–∞—à—É —Å—Ç—Ä–æ–∫—É: "))
def f(a):
    for i in range(len(a)):
        if a[:i] + a[i+1:] == (a[:i] + a[i+1:])[::-1]:
            return 1
    return 0
if f(a) == 1:
    print(f"–°—Ç—Ä–æ–∫–∞ '{a}' —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º")
else:
    print(f"–°—Ç—Ä–æ–∫–∞ '{a}' –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º")
```

4. –°–¥–µ–ª–∞–π—Ç–µ —Ç–∞–∫, —á—Ç–æ–±—ã —á–∏—Å–ª–æ —Å–µ–∫—É–Ω–¥ –æ—Ç–æ–±—Ä–∞–∂–∞–ª–æ—Å—å –≤
–≤–∏–¥–µ–¥–Ω–∏:—á–∞—Å—ã:–º–∏–Ω—É—Ç—ã:—Å–µ–∫—É–Ω–¥—ã.

```python
s = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ —Å–µ–∫—É–Ω–¥: "))

def time(s):
    d = s // 86400
    s %= 86400
    h = s // 3600
    s %= 3600
    m = s // 60
    s %= 60
    
    print(f"{d} - –¥: {h} - —á: {m} - –º: {s} - —Å")

time(s)
```
